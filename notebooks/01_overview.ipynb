# 01 – Data Overview

Quick exploration of the raw REFIT data for the 5 selected houses: **2, 3, 4, 5, 8**.


```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path().resolve().parent))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.load_refit import load_config, load_house_raw

cfg = load_config('../configs/config.yaml')
HOUSES = cfg['houses']          # [2, 3, 4, 5, 8]
RAW_DIR = '../' + cfg['data']['raw_dir']
print('Houses:', HOUSES)
```

```python
# Load one house and display basic stats
house_id = HOUSES[0]
df = load_house_raw(house_id, RAW_DIR, cfg['house_channels'][house_id])
print(df.shape)
df.head()
```

```python
# Missing data summary
missing = df.isna().mean() * 100
missing.plot.bar(title=f'House {house_id} – % missing values')
plt.ylabel('%')
plt.tight_layout()
plt.show()
```

```python
# Plot aggregate + appliances for 3 days
sample = df.iloc[:4320]   # ~3 days at 8s
fig, axes = plt.subplots(len(df.columns), 1, figsize=(14, 2*len(df.columns)), sharex=True)
for ax, col in zip(axes, df.columns):
    ax.plot(sample[col], lw=0.6)
    ax.set_ylabel(col, fontsize=8)
plt.suptitle(f'House {house_id} – raw power (3 days)', y=1.01)
plt.tight_layout()
plt.show()
```
