# 02 – Cleaning & Resampling

Demonstrates the cleaning pipeline: negative removal, outlier clipping, 1-minute resampling, and limited interpolation.


```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path().resolve().parent))

import pandas as pd
import matplotlib.pyplot as plt
from src.load_refit import load_config, load_house_raw
from src.cleaning import clean_dataframe

cfg = load_config('../configs/config.yaml')
house_id = cfg['houses'][0]
RAW_DIR = '../' + cfg['data']['raw_dir']

df_raw = load_house_raw(house_id, RAW_DIR, cfg['house_channels'][house_id])
```

```python
c = cfg['cleaning']
df_clean, report = clean_dataframe(
    df_raw,
    resample_freq=cfg['resample_freq'],
    outlier_quantile=c['outlier_quantile'],
    interp_method=c['interp_method'],
    interp_limit=c['interp_limit'],
    max_missing_ratio=c['max_missing_ratio'],
)
print(df_clean.shape)
import pandas as pd; pd.DataFrame(report).T
```

```python
# Before / After for one appliance
app = 'kettle'
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 5), sharex=False)
df_raw[app].iloc[:500].plot(ax=ax1, title=f'{app} – raw', lw=0.7)
df_clean[app].iloc[:500].plot(ax=ax2, title=f'{app} – cleaned & 1-min', color='orange', lw=0.7)
plt.tight_layout()
plt.show()
```
