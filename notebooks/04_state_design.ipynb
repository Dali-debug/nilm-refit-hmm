# 04 – State Design

Visualise power distributions and choose HMM state counts per appliance.

```python
import sys
sys.path.insert(0, '..')
import matplotlib.pyplot as plt
from src.config import DEV_HOUSES, TARGET_APPLIANCES, N_STATES
from src.io_refit import load_house
from src.mapping import standardize
from src.cleaning import clean_resample
from src.states import build_state_definitions
from src.plots import plot_state_histogram
```

```python
house_id = DEV_HOUSES[0]
raw = load_house(house_id, raw_dir='../data/raw')
std = standardize(raw, house_id, appliances=TARGET_APPLIANCES)
clean, _ = clean_resample(std)
```

```python
for app in TARGET_APPLIANCES:
    if app not in clean.columns:
        continue
    defs = build_state_definitions(clean[app], app, n_states=N_STATES[app])
    fig = plot_state_histogram(clean[app], defs, appliance=app)
    plt.show()
    print(f'{app}: means={defs["means"].round(1)}, labels={defs["labels"]}')
```
