# 05 – LOHO Evaluation

Run the evaluation script and display KPI tables and plots.


```python
import subprocess, sys
result = subprocess.run(
    [sys.executable, '../scripts/evaluate_loho.py', '--config', '../configs/config.yaml'],
    capture_output=True, text=True
)
print(result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

summary = pd.read_csv('../results/summary.csv')
summary
```

```python
# F1 per appliance (box plot across LOHO folds)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=summary, x='appliance', y='f1', ax=axes[0])
axes[0].set_title('F1 (ON/OFF) per appliance – LOHO')
axes[0].set_ylim(0, 1)

sns.boxplot(data=summary, x='appliance', y='nrmse', ax=axes[1])
axes[1].set_title('NRMSE per appliance – LOHO')

plt.tight_layout()
plt.show()
```

```python
# Mean ± std table
summary.groupby('appliance')[['mae', 'rmse', 'nrmse', 'ree', 'f1']].agg(['mean', 'std']).round(3)
```
