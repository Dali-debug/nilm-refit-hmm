# 05 – Train HMM Cross-House (LOHO)

Train HMMs in Leave-One-House-Out mode using processed data.

```python
import sys
sys.path.insert(0, '..')
import pandas as pd
from pathlib import Path
from src.config import FULL_HOUSES, TARGET_APPLIANCES, PROCESSED_DATA_DIR, MODELS_DIR
from src.hmm_train import train_all_appliances
```

```python
processed_dir = Path('..') / PROCESSED_DATA_DIR
house_data = {}
for hid in FULL_HOUSES:
    p = processed_dir / f'house_{hid}.parquet'
    if p.exists():
        house_data[hid] = pd.read_parquet(p)
        print(f'House {hid}: {len(house_data[hid])} rows')
    else:
        print(f'House {hid}: not found (run preprocess_all.py)')
```

```python
if len(house_data) >= 2:
    available = list(house_data.keys())
    test_house = available[0]
    train_houses = available[1:]
    models = train_all_appliances(
        house_data=house_data,
        train_houses=train_houses,
        appliances=TARGET_APPLIANCES,
        models_dir=str(Path('..') / MODELS_DIR),
        test_house=test_house,
        n_iter=50,
    )
    print('Trained models:', list(models.keys()))
else:
    print('Not enough houses processed. Run scripts/preprocess_all.py first.')
```
